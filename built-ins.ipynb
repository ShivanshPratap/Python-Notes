int(): Returns an integer object or converts a string or number to an integer.

license(): Prints the Python software license details.

print(): Prints output to the console.
property(): Gets, sets, or deletes a property of an object.
range(): Returns a sequence of numbers within a specified range.
repr(): Returns a string containing a printable representation of an object.
reversed(): Returns a reversed iterator of a sequence.
round(): Rounds a number to a specified number of decimal places.
runfile(): Executes a file as a Python script.
set(): Returns a new set object.

slice(): Returns a slice object representing a specified range of indices.
sorted(): Returns a sorted version of a sequence or an iterable.
staticmethod(): Returns a static method for a function.
str(): Returns a string version of an object.
sum(): Returns the sum of all items in a sequence or an iterable.
super(): Returns a temporary object of the superclass.
tuple(): Returns a new tuple object.
type(): Returns the type of an object.
vars(): Returns the __dict__ attribute of an object.
